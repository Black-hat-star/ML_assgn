```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Create a sample dataset
data = {'Income': [50000, 75000, 100000, 60000, 120000]}
df = pd.DataFrame(data)

# Instantiate the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the 'Income' column
df['Income_scaled'] = scaler.fit_transform(df[['Income']])

# Display the original and scaled data
print(df)
```

```python
import pandas as pd
from sklearn.preprocessing import Normalizer

# Create a sample dataset
data = {'Income': [50000, 75000, 100000, 60000, 120000]}
df = pd.DataFrame(data)

# Instantiate the Normalizer
normalizer = Normalizer(norm='l2')  # 'l2' is the default norm for Euclidean norm

# Fit and transform the 'Income' column
df['Income_normalized'] = normalizer.transform(df[['Income']])

# Display the original and normalized data
print(df)
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Standardize the data
X_standardized = StandardScaler().fit_transform(X)

# Apply PCA with 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_standardized)

# Create a DataFrame with the PCA results
df_pca = pd.DataFrame(data=X_pca, columns=['Principal Component 1', 'Principal Component 2'])
df_pca['Target'] = y

# Visualize the 2D representation of the data
plt.figure(figsize=(8, 6))
targets = [0, 1, 2]
colors = ['r', 'g', 'b']

for target, color in zip(targets, colors):
    indices = df_pca['Target'] == target
    plt.scatter(df_pca.loc[indices, 'Principal Component 1'],
                df_pca.loc[indices, 'Principal Component 2'],
                c=color,
                label=f'Target {target}')

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.title('PCA of Iris Dataset')
plt.show()
```

```python
import numpy as np

# Given dataset
data = np.array([1, 5, 10, 15, 20])

# Define the new min and max values
new_min = -1
new_max = 1

# Min-Max scaling
scaled_data = (data - np.min(data)) / (np.max(data) - np.min(data)) * (new_max - new_min) + new_min

print("Original data:", data)
print("Scaled data (-1 to 1):", scaled_data)
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Sample dataset
data = np.array([
    [170, 65, 30, 0, 120],
    [160, 55, 25, 1, 130],
    [180, 75, 35, 0, 110],
    [165, 60, 28, 1, 125],
    [175, 70, 32, 0, 115],
])

# Separate features and labels
X = data[:, :-1]

# Standardize the data
X_standardized = StandardScaler().fit_transform(X)

# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_standardized)

# Examine explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Print explained variance ratio for each principal component
print("Explained Variance Ratio:", explained_variance_ratio)

# Calculate cumulative explained variance
cumulative_variance = np.cumsum(explained_variance_ratio)

# Print cumulative explained variance
print("Cumulative Explained Variance:", cumulative_variance)
```
